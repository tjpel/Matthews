```python
df['Owner Phone'] = df['Owner Phone'].astype(str)
df['Recorded Owner Phone'] = df['Recorded Owner Phone'].astype(str)
df = df.rename(columns = {"RBA": "Rentable Building Area"})
df.head()
```

```python
#building a dataframe for linear regression

mldf = df[['Number Of Units', 'Star Rating', 'Building Class', 'City', 'Vacancy %', 'Year Built', 'Number Of Stories', 'Land Area (AC)', 'Rentable Building Area', 'Last Sale Price']]
mldf.dropna(axis=0, inplace=True)
mldf.head()
```

```python
mldf = pd.get_dummies(mldf, columns=['City', "Building Class"])
mldf.head()
```

```python
target = mldf['Last Sale Price']
features = mldf.loc[:, mldf.columns != "Last Sale Price"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

y_hat = lr.predict(x_test)
mae = mean_absolute_error(y_test, y_hat) #how much am I off on average?
```

```python
#add values

test_total = pd.concat([x_test, y_test], axis=1)
test_total["Predicted Price"] = y_hat
test_total["Difference"] = test_total["Last Sale Price"] - test_total["Predicted Price"]
test_total["Difference %"] = test_total["Difference"] / test_total["Last Sale Price"]
test_total.head()
```

```python
test_total_pos = test_total[test_total["Difference"] <= -1*mae] #find properties that are selling for less than their worth, with a difference above our mean absolute error
len(test_total_pos)
```
